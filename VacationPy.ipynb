### Store Part I results into DataFrame

```python
# load in the csv to a dataframe 
weather_df = pd.read_csv("Data/weather_data.csv")
del weather_df["Unnamed: 0"]
weather_df
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# access maps with api_key 
gmaps.configure(api_key=g_key)
```

```python
# store lat and lng into locations
locations = weather_df[["Lat", "Lng"]]
# store humidity into variable
city_humidity = weather_df["Humidity"]

# create the humidity heatmap 
humid_fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=city_humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 3)
humid_fig.add_layer(heat_layer)

humid_fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#hotel_df = hotel_df.insert(9,"Hotel Name", "")
new_locations = hotel_df[["Lat", "Lng"]]
hotel_df
```

```python
# search for hotel within 5000 meters 
hotel_search = "Hotel"
hotel_radius = 5000
hotel_type = "lodging"

start = 0
for index, row in hotel_df.iterrows():
    # set up parameters
    params = {
        "location": locations,
        "keyword": hotel_search,
        "radius": hotel_radius,
        "type": hotel_type,
        "key": g_key
    }

    # base url
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # perform a request with new parameters 
    response = requests.get(base_url, params=params)
    response_json = response.json()
    results = response_json['results']
    if len(results) > 0:
        name = response_json['results'][0]['name']
        hotel_df.iloc[start, 9] = name
    start = start + 1
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
humid_fig.add_layer(marker_layer)

# Display Map
humid_fig
```
